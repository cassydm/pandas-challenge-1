## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty

#creating the new column that multiplies price and qty
df["line_subtotal"]=df["unit_price"]*df["qty"]
#creating a new dataframe with specific columns for easier viewing
subtotal_df = df[["client_id","unit_price","qty", "line_subtotal"]]
subtotal_df.head()
```

```python
#creating a new function to check total weight
def ship_price(x,y):
    """Calculate the shipping price based on the product of unit weight and quantity ordered.
    The shipping price is:
        $7 per pound for orders over 50lbs total (unit weight * qty)
        $10 per pound for orders under or equal to 50lbs total (unit weight *qty)

    Parameters:
    x (float): The unit weight of the item.
    y (int): The quantity of the item.

    Returns:
    float: The calculated shipping price.
    """
    if x*y > 50: 
        return 7*(x*y)
    else: 
        return 10*(x*y)
```

```python
#creating a function that applies the total shipping price (ship_price function)
#to rows in a data frame
def calc_ship_price(row):
    """Calculate the shipping price based on the unit weight and quantity of an item.

    This function takes a row from a DataFrame as input, extracts the unit weight and quantity
    of the item from the specified columns, and calculates the shipping price using the ship_price function.

    Parameters:
    row (pandas.Series): A row from a DataFrame containing the unit weight and quantity of an item.

    Returns:
    float: The calculated shipping price.
    """
    return ship_price(row["unit_weight"],row["qty"])
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df["total_weight"]=df["qty"]*df["unit_weight"]
df["shipping_price"]=df.apply(calc_ship_price,axis=1)

#creating a new data frame that shows only nescessary information
ship_price_df=df[["client_id","unit_price","unit_weight","qty","total_weight","shipping_price"]]
ship_price_df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
#defining the tax rate for multiplying
tax=1.0925

df["line_price"]=(df["shipping_price"]+df["line_subtotal"])*(tax)
#creating a new dataframe showing only this info
pricing_df=df[["line_subtotal","shipping_price","line_price"]]
pricing_df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df["line_cost"]=(df["unit_cost"]*df["qty"])+df["shipping_price"]
df.head(3)
```

```python
# Create a column for the profit of each line using line cost and line price
df["profit"]=df["line_price"]-df["line_cost"]
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

#calling the list of the top 5 clients
top5_clients_list

#creating a function that filters the data to the client id and returns the sum
def client_spending_func(x):
    """Calculate the total spending of a client based on their client ID.

    This function filters the DataFrame `df` to include only the rows where the 
    `client_id` matches the provided value `x`. It then calculates and returns 
    the total spending of that client by summing the values in the `line_price` column.

    Parameters:
    x (int): The client ID to filter the DataFrame.

    Returns:
    float: The total spending of the client with the specified ID.

    """
    #filtering df as a new df based on when the client id = x
    client_spending_df=df[df["client_id"]==x]
    #finding the total spent for that filter
    total_client_sum=client_spending_df["line_price"].sum()
    return total_client_sum
 
```

```python
#creating a loop to check all clients total spending
#defining a list to store the total spending values
total_client_sum_list = []

for client_id in top5_clients_list:
    #calling client spending function to pass in client id
    total_client_sum = client_spending_func(client_id)
    #adding to the list
    total_client_sum_list.append(total_client_sum)
    print(f"Client {client_id} has spent ${total_client_sum:,.2f}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

#creating a summarized dataframe grouped by client and showing the above metrics summed
summarized_df=df.groupby("client_id").agg({
    "qty":"sum",
    "shipping_price":"sum",
    "line_price":"sum",
    "line_cost":"sum",
    "profit":"sum"
    }).reset_index()

#filtering the data to the top 5 clients
summarized_df=summarized_df[summarized_df["client_id"].isin(top5_clients_list)]

#reset index
summarized_df.reset_index(drop=True, inplace=True)
summarized_df
```

```python
# Format the data and rename the columns to names suitable for presentation.
summarized_df=summarized_df.rename(columns={
    "client_id": "Client ID",
    "qty": "Total Units",
    "shipping_price": "Shipping",
    "line_price":"Total Revenue",
    "line_cost": "Total Cost",
    "profit": "Total Profit"
})
summarized_df
```

```python

# Define the money columns. 
money_col=["Shipping","Total Revenue","Total Cost","Total Profit"]
```

```python
# Define a function that converts a dollar amount to millions.
def conversion(x):
    """
    Convert a dollar amount to millions.

    This function takes a numerical value representing a dollar amount
    and divides it by 1,000,000 to convert it to millions.

    Parameters:
    x (float or int): The dollar amount to be converted.

    Returns:
    float: The converted amount in millions.
    """
    #divide by 1m and return
    return x/1000000
```

```python
# Apply the currency_format_millions function to only the money columns. 

for col in money_col:
    #using the function to convert the money cols
    summarized_df[col]=summarized_df[col].apply(conversion)

summarized_df
```

```python
# Rename the columns to reflect the change in the money format. 
summarized_df=summarized_df.rename(columns={
    "Shipping": "Shipping (millions)",
    "Total Revenue":"Total Revenue (millions)",
    "Total Cost": "Total Cost (millions)",
    "Total Profit": "Total Profit (millions)"
})

summarized_df
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sorted_df=summarized_df.sort_values(["Total Profit (millions)"], ascending=False)
sorted_df
```
